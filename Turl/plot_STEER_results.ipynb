```python
import json
import matplotlib.pyplot as plt
import numpy as np
from os.path import join

corpus = "turl"
x_values = [1,2,3,4,5]
avg = "weighted"
metric = "f1-score"

####### TURL Baseline
f1_values_baseline = []

for labeled_data_size in x_values:
    with open(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0_2/classification_report.json") as f:
        results = json.load(f)
        f1_values_baseline.append(results[f"{avg} avg"][metric])


###### TURL+STEER
f1_values_STEER = []
for labeled_data_size in x_values:
    with open(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0_2_STEER/classification_report.json") as f:
        results = json.load(f)
        f1_values_STEER.append(results[f"{avg} avg"][metric])


###### TURL+STEER no overwrite
f1_values_STEER_no_overwrite = []
for labeled_data_size in [1]:
    with open(f"output/CT/v2/1/model_STEER_1_absolute_20.0_2_no_overwrite/classification_report.json") as f:
        results = json.load(f)
        f1_values_STEER_no_overwrite.append(results[f"{avg} avg"][metric])

###### TURL 80% train
f1_values_TURL80 = []
lds_sameTraindata = []
for lds, ulds in [("14.30","65.70"),("14.56","65.44"),("14.74","65.26"),("14.86","65.14"),("14.89","65.11")]:
    # with open(f"output/CT/v2/1/model_STEER_5_absolute_20.0_2_80train/classification_report.json") as f:
    #     results = json.load(f)
    #     f1_values_TURL80.append(results[f"{avg} avg"][metric])
    with open(f"output/CT/v2/1/model_STEER_{lds}_{ulds}_20.00_2_sameTraindata/classification_report.json") as f:
        results = json.load(f)
        f1_values_TURL80.append(results[f"{avg} avg"][metric])
    lds_sameTraindata.append(lds)
    # if avg == "weighted":
    #     f1_values_TURL80.append(0.72)
    # else:
    #     f1_values_TURL80.append(0.5)


fontsize = 20
plt.figure(figsize=(9,7))
#plt.errorbar(x_values, y_values_baseline, error_values_baseline, marker="o", color="C0", capsize=3, label="Sato baseline", linewidth=3)
plt.plot(x_values, f1_values_baseline, marker="o",  color="C1", label="Turl retrain", linewidth=3)

#plt.errorbar(x_values, y_values_weakSup, error_values_weakSup, marker="o",capsize=3, color="C3", label="LFBasedLabeler", linewidth=3)
plt.plot(x_values, f1_values_STEER, marker="o", color="C2", label="STEER", linewidth=3)
#plt.plot(x_values, f1_values_TURL80, color="red", marker="o", label="same amount of train data (STEER, sampled)", linewidth=3)
#plt.plot([1], f1_values_STEER_no_overwrite, color="red", marker="o", label="same STEER train data (no overwrite)", linewidth=3)


plt.yticks(np.arange(0,0.7,0.1), fontsize=fontsize)
plt.xticks(np.arange(1,6,1), fontsize=fontsize)
plt.grid()
plt.legend(loc="upper center", bbox_to_anchor=(0.5,1.15), ncol=2, fancybox=True,shadow=True, fontsize=fontsize)
#plt.title(f"Public BI ({avg})", fontsize=20)
plt.xlabel("labeled data size: # of labeled cols per sem. type", fontsize=fontsize)
#plt.xlabel("labeled data size: # of labeled cols per sem. type\n Dataset: TURL", fontsize=fontsize)
plt.ylabel(f"{avg} {metric}", fontsize=fontsize)
plt.savefig(join("plots", f"turl_{corpus}_{avg}_{metric}.png"), bbox_inches='tight', dpi=100)
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os.path import join

corpus = "public_bi"

x_values = [1,2,3,4,5]
avg = "weighted"
metric = "f1-score"

####### TURL Baseline
f1_values_baseline = []
error_values_baseline = []

for labeled_data_size in x_values:
    # with open(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0_2_PublicBI/classification_report.json") as f:
    #     results = json.load(f)
    #     f1_values_baseline.append(results[f"{avg} avg"][metric])
    df_result = pd.read_csv(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0PublicBI_mean.csv", index_col=0)
    f1_values_baseline.append(df_result.loc[f"{metric}s_{avg}"]["mean"])
    error_values_baseline.append(df_result.loc[f"{metric}s_{avg}"]["std"])

###### TURL+STEER
f1_values_STEER = []
error_values_STEER = []
for labeled_data_size in x_values:
    # with open(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0_2_PublicBI_STEER/classification_report.json") as f:
    #     results = json.load(f)
    #     f1_values_STEER.append(results[f"{avg} avg"][metric])
    df_result = pd.read_csv(f"output/CT/v2/1/model_STEER_{labeled_data_size}_absolute_20.0PublicBI_STEER_mean.csv", index_col=0)
    f1_values_STEER.append(df_result.loc[f"{metric}s_{avg}"]["mean"])
    error_values_STEER.append(df_result.loc[f"{metric}s_{avg}"]["std"])

###### TURL 80% train
f1_values_TURL80 = []
lds_sameTraindata = []
for labeled_data_size, unlabeled_data_size in [("43.19","36.80"),("50.84","29.14"),("55.06","24.93"),("59.27","20.72"),("61.24","18.75")]:
    with open(f"output/CT/v2/1/model_STEER_{labeled_data_size}_{unlabeled_data_size}_20.0_2_PublicBI_sameTraindata/classification_report.json") as f:
        results = json.load(f)
        f1_values_TURL80.append(results[f"{avg} avg"][metric])
    lds_sameTraindata.append(labeled_data_size)
    # if avg == "weighted":
    #     f1_values_TURL80.append(0.55)
    # else:
    #     f1_values_TURL80.append(0.25)


fontsize = 20
plt.figure(figsize=(9,7))
plt.errorbar(x_values, f1_values_baseline, error_values_baseline, marker="o", color="C1", capsize=3, label="Turl retrain", linewidth=3)
#plt.plot(x_values, f1_values_baseline, marker="o",  color="C1", label="Turl retrain", linewidth=3)

#plt.errorbar(x_values, y_values_weakSup, error_values_weakSup, marker="o",capsize=3, color="C3", label="LFBasedLabeler", linewidth=3)
plt.errorbar(x_values, f1_values_STEER, error_values_STEER, marker="o", capsize=3, color="C2", label="STEER", linewidth=3)
#plt.plot(x_values, f1_values_TURL80, color="red", marker="o", label="same amount of train data (STEER, sampled)", linewidth=3)

plt.yticks(np.arange(0,0.7,0.1), fontsize=fontsize)
plt.xticks(np.arange(1,6,1), fontsize=fontsize)
plt.grid()
plt.legend(loc="upper center", bbox_to_anchor=(0.5,1.15), ncol=2, fancybox=True,shadow=True, fontsize=fontsize)
#plt.title(f"Dataset: Public BI ({avg})", fontsize=20)
plt.xlabel("labeled data size: # of labeled cols per sem. type", fontsize=fontsize)
#plt.xlabel("labeled data size: # of labeled cols per sem. type\n Dataset: Public BI", fontsize=fontsize)
plt.ylabel(f"{avg} {metric}", fontsize=fontsize)
plt.savefig(join("plots", f"turl_{corpus}_{avg}_{metric}.png"), bbox_inches='tight', dpi=100)
plt.show()
```
