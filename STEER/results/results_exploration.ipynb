```python
import os
from os.path import join
from dotenv import load_dotenv
load_dotenv(override=True)
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import json
from sklearn.preprocessing import LabelEncoder

os.environ["TYPENAME"] = "type_public_bi"

# LabelEncoder
with open(
        join(os.environ["WORKING_DIR"], "data", "extract", "out",
             "valid_types", "types.json")) as f:
    valid_types = json.load(f)[os.environ["TYPENAME"]]

label_enc = LabelEncoder()
label_enc.fit(valid_types)

corpus = "public_bi_num"
metric = "f1-scores"
avg = "weighted"



x_values = [1,2,3,4,5]
random_state = 2

class_reports = []
########## Sato retrained with already labeled data

for labeled_data_size in x_values:
    # get Sato Baseline scores
    with open(join(os.environ["SATO_DIR"], "results", "CRF_log", os.environ["TYPENAME"], f"CRF+LDA_eval_eval_labeled{labeled_data_size}_unlabeledAbsolute_test20.0_{random_state}", "outputs", f"classification_report_CRF+LDA_retrain_labeled{labeled_data_size}_unlabeledAbsolute_test20.0_{random_state}_train100.pt.json")) as f:
        class_report = json.load(f)
    new_class_report = {}
    for key in class_report.keys():
        try:
            type_int = int(key)
            sem_type = label_enc.inverse_transform([type_int])[0]
            new_class_report[sem_type] = class_report[key]
        except:
            new_class_report[key] = class_report[key]
    class_reports.append(new_class_report)

class_reports_STEER = []
########## Sato retrained with STEER
for labeled_data_size in x_values:
    # get Sato Baseline scores
    with open(join(os.environ["SATO_DIR"], "results", "CRF_log", os.environ["TYPENAME"], f"CRF+LDA_eval_eval_embclus_all_combined_maj_labeled{labeled_data_size}_unlabeledAbsolute_test20.0_{random_state}", "outputs", f"classification_report_CRF+LDA_retrain_embclus_all_combined_maj_labeled{labeled_data_size}_unlabeledAbsolute_test20.0_{random_state}_train100.pt.json")) as f:
        class_report = json.load(f)
    new_class_report = {}
    for key in class_report.keys():
        try:
            type_int = int(key)
            sem_type = label_enc.inverse_transform([type_int])[0]
            new_class_report[sem_type] = class_report[key]
        except:
            new_class_report[key] = class_report[key]
    class_reports_STEER.append(new_class_report)
            
## Plot
fig, ax = plt.subplots(figsize=(20,10))
metric = "f1-score"

width = 1.5

scores_per_label = []
#for class_report in class_reports_STEER:#class_reports:
for class_report in class_reports:
    scores = []
    for key in class_report.keys():
        if key == "accuracy":
            scores.append([key,class_report[key]])
        elif key in ["macro avg","weighted avg"]:
            scores.append([key,class_report[key][f"{metric}"]])
        else:
            if class_report[key]["support"] > 0:
                scores.append([key, class_report[key][f"{metric}"]])
    scores_per_label.append(scores)


X = np.arange(0,len([x[0] for x in scores_per_label[0]])*10,10)
ax.bar(X - (2*width), [x[1] for x in scores_per_label[0]], width=width, label="labeled_data_size 1")
ax.bar(X - (1*width), [x[1] for x in scores_per_label[1]], width=width, label="labeled_data_size 2")
ax.bar(X, [x[1] for x in scores_per_label[2]], width=width, label="labeled_data_size 3")
ax.bar(X + (1*width), [x[1] for x in scores_per_label[3]], width=width, label="labeled_data_size 4")
ax.bar(X + (2*width), [x[1] for x in scores_per_label[4]], width=width, label="labeled_data_size 5")
ax.set_xticks(X)
ax.set_ylabel(f"{metric}")
ax.set_yticks(np.arange(0,1.1,0.1))
ax.set_xticklabels([x[0] for x in scores_per_label[0]], rotation=90)

ax.grid(axis="y")
ax.legend(loc=0)
ax.set_title(f"Scores per semantic type: {metric}")

fig.tight_layout()

plt.show()
```
